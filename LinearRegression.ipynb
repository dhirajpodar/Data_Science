```python
import numpy as np
import pandas as pd

import  matplotlib.pyplot as plt
```

```python
pages_speed = np.random.normal(3.0,1.0,1000)
purchase_amount = 100 - (pages_speed + np.random.normal(0,1.0,1000)) * 3

plt.scatter(pages_speed,purchase_amount)
```

```python
from scipy import stats

slope, intercept,r_value,p_value,std_err = stats.linregress(pages_speed,purchase_amount)

r_value ** 2
```

```python
def predict(x):
    return slope * x + intercept

fitLine = predict(pages_speed)

plt.scatter(pages_speed,purchase_amount)
plt.plot(pages_speed,fitLine,c='r')
plt.show()
```
